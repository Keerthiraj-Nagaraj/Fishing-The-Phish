```python
def plot_history(history):
    acc = history.history['acc']
    val_acc = history.history['val_acc']
    loss = history.history['loss']
    val_loss = history.history['val_loss']
    x = range(1, len(acc) + 1)

    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    plt.plot(x, acc, 'b', label='Training acc')
    plt.plot(x, val_acc, 'r', label='Validation acc')
    plt.title('Training and validation accuracy')
    plt.legend()
    plt.subplot(1, 2, 2)
    plt.plot(x, loss, 'b', label='Training loss')
    plt.plot(x, val_loss, 'r', label='Validation loss')
    plt.title('Training and validation loss')
    plt.legend()
```

```python
def text_processor(text):
    
    ## Convert words to lower case and split them
    text = text.lower().split()
    
    ## Remove stop words
    stop_words = set(stopwords.words("english"))
    text = [w for w in text if not w in stop_words]  
    
    text = " ".join(text)

    ## Clean the text
    text = re.sub(r"[^A-Za-z0-9^,!.\/'+-=]", " ", text)
    text = re.sub(r"what's", "what is ", text)
    text = re.sub(r"\'s", " ", text)
    text = re.sub(r"\'ve", " have ", text)
    text = re.sub(r"n't", " not ", text)
    text = re.sub(r"i'm", "i am ", text)
    text = re.sub(r"\'re", " are ", text)
    text = re.sub(r"\'d", " would ", text)
    text = re.sub(r"\'ll", " will ", text)
    text = re.sub(r",", " ", text)
    text = re.sub(r"\.", " ", text)
    text = re.sub(r"!", " ! ", text)
    text = re.sub(r"\/", " ", text)
    text = re.sub(r"\^", " ^ ", text)
    text = re.sub(r"\+", " + ", text)
    text = re.sub(r"\-", " - ", text)
    text = re.sub(r"\=", " = ", text)
    text = re.sub(r"'", " ", text)
    text = re.sub(r"(\d+)(k)", r"\g<1>000", text)
    text = re.sub(r":", " : ", text)
    text = re.sub(r" e g ", " eg ", text)
    text = re.sub(r" u s ", " american ", text)
    text = re.sub(r"\0s", "0", text)
    text = re.sub(r"e - mail", "email", text)
    text = re.sub(r"\s{2,}", " ", text)
    
 
    # Tokenize
    tokens = nltk.word_tokenize(text)
  
    # Lemmatize 

    lemmatizer = WordNetLemmatizer()
    lemmatized_words = [lemmatizer.lemmatize(word) for word in tokens]
    
    lemmatized_words = [w for w in lemmatized_words if len(w)>1]  
    
    text = " ".join(lemmatized_words)
    return text
```

```python
X_train, X_test, y_train, y_test = train_test_split(num_data_scaled, num_lab, test_size = 0.2, random_state = 0)
```

```python
names = ["Log-Reg", "Nearest Neighbors",
         "Decision Tree", "Random Forest", "AdaBoost"]

classifiers = [
    LogisticRegression(),
    KNeighborsClassifier(5),
    DecisionTreeClassifier(max_depth=5),
    RandomForestClassifier(max_depth=5, criterion = 'gini', max_features = 'log2', n_estimators = 50),
    AdaBoostClassifier()]
```

```python
y_pred_mat_num = np.zeros((len(X_test), len(names)))
```

```python
f1_vals = []

i = 0

for name, clf in zip(names, classifiers):
    
    print(name)
    clf.fit(X_train, y_train)
    score = clf.score(X_test, y_test)
    y_hat = clf.predict(X_test)
    
    y_pred_mat_num[:,i] = y_hat
    
    f1_vals.append(f1_score(y_test, y_hat))
    
    target_names = ["Not Phishing", "Phishing"]
    print(classification_report(y_test, y_hat, target_names=target_names))
    print(confusion_matrix(y_test, y_hat))
    print("Done")
    
    i += 1
    
print("Done")
    
```

```python
x = np.arange(0,len(f1_vals))

plt.plot(x, f1_vals)

# You can specify a rotation for the tick labels in degrees or with keywords.
plt.xticks(x, names, rotation='vertical')

print(np.max(f1_vals))
```

```python
X_train_text, X_test_text, y_train_text, y_test_text = train_test_split(clean_url, num_lab, test_size = 0.2, random_state = 0)
```

```python
y_pred_mat_text = np.zeros((len(X_test_text), len(names)))
```

```python
f1_vals_text = []

i = 0

for name, clf in zip(names, classifiers):
    
    print(name)
    
    classifier = Pipeline([('vect', CountVectorizer()), ('tfidf', TfidfTransformer()), ('clf', clf)])
    
    classifier.fit(X_train_text, y_train_text)
    y_hat_text = classifier.predict(X_test_text)
    
    y_pred_mat_text[:,i] = y_hat_text
    
    f1_vals_text.append(f1_score(y_test_text, y_hat_text))
    
    #target_names = ["Not Phishing", "Phishing"]
    #print(classification_report(y_test_text, y_hat_text, target_names=target_names))
    
    print(confusion_matrix(y_test_text, y_hat_text))
    
    i += 1
    print("Done")
    
print("Done")
```

```python
x = np.arange(0,len(f1_vals_text))

plt.plot(x, f1_vals_text)

# You can specify a rotation for the tick labels in degrees or with keywords.
plt.xticks(x, names, rotation='vertical')

print(np.max(f1_vals_text))
```

```python
nn = y_pred_mat_text[:,1] + y_pred_mat_num[:,1]
```

```python
nn_pred = []

for val in nn:
    if val > 1:
        nn_pred.append(1)
    else:
        nn_pred.append(0)
        
nn_pred = np.array(nn_pred)

confusion_matrix(y_test_text, nn_pred)

print(f1_score(y_test_text, nn_pred))
```
